```python
import pandas as pd
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import numpy as np
import math
from pandas.core import datetools
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf, pacf
import statsmodels.tsa.stattools as ts
from statsmodels.tsa.arima_model import ARIMA

```

```python
df = pd.read_csv('/Users/ujash/Documents/IlabDatasets/BHP.AX_weekly.csv')
df.head() 

df.index=pd.to_datetime(df['Date'])
df['Close'].plot()
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df['Close'].diff().dropna(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df['Close'].diff().dropna(), lags=40, ax=ax2)
plt.show()
```

```python
model=sm.tsa.statespace.SARIMAX(endog=df['Close'],order=(0,1,0),seasonal_order=(0,0,1,6),trend='c',enforce_invertibility=False)
results=model.fit()
print(results.summary())
```

```python
#to print the first diff and then seasonal difference which provide us almost zero
print(df['Close'].diff().diff(12))
```

```python
#if we just want to print seasonal difference
print(df['Close'].diff(12))
```

```python
# as the growth rate is zero, we take here new random dataset (noise), generate an AR(1) process, with noise having a $2 standard deviation
np.random.seed(120)
noise=[np.random.normal(scale=2)]


for i in range(len(df)-1):
    noise.append(np.random.normal(scale=2)+noise[i]*(-0.85))
df['Close2']=df['Close']+noise
df['Close2'].plot()
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df['Close2'].diff().diff(12).dropna(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df['Close2'].diff().diff(12).dropna(), lags=40, ax=ax2)
plt.show()

print(sm.tsa.stattools.adfuller(df['Close2'].diff().diff(12).dropna()))
```

```python
model=sm.tsa.statespace.SARIMAX(endog=df['Close2'],order=(1,1,0),seasonal_order=(0,1,0,12),trend='c',enforce_invertibility=False)
results=model.fit()
print(results.summary())
```

```python
model2=sm.tsa.statespace.SARIMAX(endog=df['Close2'],order=(1,1,0),seasonal_order=(0,1,1,12),trend='c',enforce_invertibility=False)
results2=model2.fit()
print(results2.summary())
```

```python
df['noise']=[noise[i]+0.85*noise[i-1] if i>0 else 0 for i in range(len(noise))]
results2.resid.loc['1999-03-01':].plot(label='Regression Residuals')
df['noise'].loc['1999-03-01':].plot(color='r',label='True Noise')
plt.legend(loc=2)
plt.show()
```
